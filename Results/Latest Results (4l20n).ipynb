## Calculate $L_2$ error

```python
import pickle
import numpy as np

Total_error_u_test_num = 0
Total_error_u_test_den = 0

for j in range(1,5):
    file = open('Ht_4_l_20_n_%d_fold_constant_fitsolver_rep_v(with_train)'%(j),'rb')
    data = pickle.load(file)
    file.close()
    
    u_test_tot = data['u_test_tot']
    u_test_pred = data['u_test_pred']
    K_test_tot = data['K_test_tot']
    alpha_opt = data['alpha']
    
    error_u_test_num = (np.linalg.norm(u_test_tot*K_test_tot-u_test_pred*K_test_tot))**2
    error_u_test_den = (np.linalg.norm(u_test_tot*K_test_tot))**2
    print('%.4e'%np.sqrt(error_u_test_num/error_u_test_den))
    print('alpha_opt = %s'%alpha_opt)
    print('\n')
    Total_error_u_test_num += error_u_test_num
    Total_error_u_test_den += error_u_test_den
    
print('%.4e'%np.sqrt(Total_error_u_test_num/Total_error_u_test_den))

```

## *Plot*: 2D line plots

```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
file = open('Heston_4_l_20_n_1_constant_fitsolver_alpha_3_rep_v(with_train)','rb')
data = pickle.load(file)
file.close()

X_test_tot = data['X_test_tot']
u_test_pred = data['u_test_pred']
u_test_tot = data['u_test_tot']
K_test_tot = data['K_test_tot']

X_train_tot = data['X_train_tot']
u_train_pred = data['u_train_pred']
u_train_tot = data['u_train_tot']
K_train_tot = data['K_train_tot']


test_tot = np.concatenate((X_test_tot,u_test_pred, u_test_tot, K_test_tot),axis=1)
train_tot = np.concatenate((X_train_tot,u_train_pred, u_train_tot, K_train_tot),axis=1)
tot = np.concatenate([test_tot,train_tot])

# Filter particular time
tot_t = tot[tot[:,1] == np.unique(tot[:,1])[0]]

# filter by particular v
tot_v1 = tot_t[tot_t[:,2] == np.unique(tot_t[:,2])[0]]
tot_v2 = tot_t[tot_t[:,2] == np.unique(tot_t[:,2])[20]]
tot_v3 = tot_t[tot_t[:,2] == np.unique(tot_t[:,2])[50]]

#sort by first column i.e. x since we want to plot V against x
tot_v1 = tot_v1[tot_v1[:,0].argsort()]
tot_v2 = tot_v2[tot_v2[:,0].argsort()]
tot_v3 = tot_v3[tot_v3[:,0].argsort()]

# Note the subtlety here: we cannot plot V against x because x is scaled by K nor can we plot V against s because we
# already fixed s as a constant and we will end up with a straight line if we plot it. So, can only plot u against x.
plt.figure(figsize=(8,6))
plt.plot(tot_v1[:,0], tot_v1[:,4], linewidth = 5, color = 'b')
plt.plot(tot_v1[:,0], tot_v1[:,3], linewidth = 5, linestyle = '--', color = 'r')
plt.plot(tot_v2[:,0], tot_v2[:,4], linewidth = 5, color = 'b')
plt.plot(tot_v2[:,0], tot_v2[:,3], linewidth = 5, linestyle = '--', color = 'r')
plt.plot(tot_v3[:,0], tot_v3[:,4], linewidth = 5, color = 'b')
plt.plot(tot_v3[:,0], tot_v3[:,3], linewidth = 5, linestyle = '--', color = 'r')
```

```python
def newton_vol_call(S, K, T, C, r, sigma):
    
    #S: spot price
    #K: strike price
    #T: time to maturity
    #C: Call value
    #r: interest rate
    #sigma: volatility of underlying asset
    
    d1 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    
    fx = S * si.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0) - C
    
    vega = (1 / np.sqrt(2 * np.pi)) * S * np.sqrt(T) * np.exp(-(si.norm.cdf(d1, 0.0, 1.0) ** 2) * 0.5)
    
    tolerance = 0.000001
    x0 = sigma
    xnew  = x0
    xold = x0 - 1
        
    while abs(xnew - xold) > tolerance:
    
        xold = xnew
        xnew = (xnew - fx - C) / vega
        
        return abs(xnew)
```
